```python
#define variables for graph make best fit line do linear regression
x = hrpay.iloc[:, 0:1]
y = hrpay.iloc[:, 2:3]
LR = LinearRegression()
Reg = LR.fit(x,y)
y_pred = LR.predict(x)
plt.plot(x,y_pred, color = "red")
plt.scatter(x,y)
plt.xlabel("year")
plt.ylabel("hourly wage")
```

```python
#find Print the intercept, slope and R2 values. Interpret the intercept, slope and R2 value and correlation
print(Reg.intercept_)
slp=Reg.coef_
r2 = r2_score(y,y_pred)
CC = math.sqrt(r2)
print(CC)
print(r2)
print(slp)
```

This slope suggest hourly wages increased about 18 cents per year. Assuming no vacation time at all at a 40 hour work week with 52 weeks. wages should of increased about of 374.40 a year. Since this is an average we must take into account That there are some people who have higher hourly wages which drives the average up. I think this is shown by the fact that the median is usually about 10 dollars lower which is about 20800$ year less than what the best fit line suggests. I think this really shows a desparity in wages amongst hourly workers, Which is to be expected as some jobs pay more than others.
